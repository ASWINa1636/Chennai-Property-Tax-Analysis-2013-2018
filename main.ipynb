Respacing the columns

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

```python
# Basic Structure
print(df.columns)
```

```python
print(df.describe())
```

Renaming the columns name

```python
years = ["2013_14", "2014_15", "2015_16", "2016_17", "2017_18"]

records = []

for _, row in df.iterrows():
    for year in years:
        try:
            records.append({
                "city": row["city_name"],
                "zone": row["zone_name"],
                "ward": row["ward_name"],
                "ward_no": row["ward_no."],
                "year": year,

                "res_collection": row[f"{year}___property_tax_collection_in_crores___residential"],
                "res_demand": row[f"{year}___property_tax_demand_in_crores___residential"],

                "com_collection": row[f"{year}___property_tax_collection_in_crores___commercial"],
                "com_demand": row[f"{year}___property_tax_demand_in_crores___commercial"],
            })
        except KeyError:
            print(f"Column missing for {year}")
```

```python
df_long = pd.DataFrame(records)
```

```python
print(df_long.head())
```

```python
print(df_long.shape)
```

```python
print(df_long.dtypes)
```

```python
cols = ["res_collection", "res_demand", "com_collection", "com_demand"]

df_long[cols] = df_long[cols].apply(pd.to_numeric, errors='coerce')
```

```python
print(df_long.isnull().sum())
```

```python
# Safe approach
df_long.fillna(0, inplace=True)
```

## Metrics

```python
# Total metrics
df_long["total_collection"] = df_long["res_collection"] + df_long["com_collection"]
df_long["total_demand"] = df_long["res_demand"] + df_long["com_demand"]
```

```python
# Efficiency (MOST IMPORTANT KPI)
df_long["efficiency"] = df_long["total_collection"] / df_long["total_demand"]
```

```python
# Handle divide-by-zero safely
df_long["efficiency"] = df_long["efficiency"].fillna(0)
```

Year-wise Trend Analysis

```python
yearly = df_long.groupby("year")[["total_collection", "total_demand"]].sum()

yearly.plot(marker='o')
plt.title("Year-wise Demand vs Collection")
plt.ylabel("Amount (Crores)")
plt.grid()
plt.show()
```

Efficiency Trend (Critical Insight)

```python
eff_trend = df_long.groupby("year")["efficiency"].mean()

eff_trend.plot(marker='o')
plt.title("Average Efficiency Over Years")
plt.ylabel("Efficiency")
plt.grid()
plt.show()
```

Zone-Level Performance

```python
zone_perf = df_long.groupby("zone")[["total_collection", "total_demand"]].sum()

zone_perf["efficiency"] = zone_perf["total_collection"] / zone_perf["total_demand"]

zone_perf = zone_perf.sort_values(by="efficiency", ascending=False)

print(zone_perf.head(10))   # Best zones
print(zone_perf.tail(10))   # Worst zones
```

```python
ward_perf = df_long.groupby("ward_no")["efficiency"].mean().sort_values()

print("Worst Wards:\n", ward_perf.head(10))
print("Best Wards:\n", ward_perf.tail(10))
```

```python
df_long["risk_flag"] = df_long["efficiency"].apply(lambda x: 1 if x < 0.6 else 0)
```

```python
df_long["gap"] = df_long["total_demand"] - df_long["total_collection"]

total_gap = df_long["gap"].sum()

print("Total Revenue Leakage (Crores):", total_gap)
```

Core Derived Features

```python
# Already done, but ensure consistency
df_long["total_collection"] = df_long["res_collection"] + df_long["com_collection"]
df_long["total_demand"] = df_long["res_demand"] + df_long["com_demand"]

df_long["efficiency"] = df_long["total_collection"] / df_long["total_demand"]
df_long["efficiency"] = df_long["efficiency"].fillna(0)
```

Revenue Gap (Key Policy Metric)

```python
df_long["gap"] = df_long["total_demand"] - df_long["total_collection"]
```

Growth RateGrowth Rate (Time Intelligence)

```python
df_long = df_long.sort_values(["zone", "year"])

df_long["growth"] = df_long.groupby("zone")["total_collection"].pct_change()
```

Commercial Dependency

```python
df_long["commercial_share"] = df_long["com_collection"] / df_long["total_collection"]
```

Stability

```python
stability = df_long.groupby("zone")["efficiency"].std().reset_index()
stability.columns = ["zone", "efficiency_std"]
```

Risk Score

```python
df_long["risk_score"] = (
    (1 - df_long["efficiency"]) * 0.5 +
    (df_long["gap"] / df_long["total_demand"]) * 0.5
)
```

Zone Ranking System 

```python
zone_score = df_long.groupby("zone").agg({
    "efficiency": "mean",
    "growth": "mean",
    "gap": "sum"
}).reset_index()

zone_score["score"] = (
    zone_score["efficiency"] * 0.5 +
    zone_score["growth"].fillna(0) * 0.3 -
    (zone_score["gap"] / zone_score["gap"].max()) * 0.2
)

zone_score = zone_score.sort_values(by="score", ascending=False)
```

1. Zone Ranking Bar Chart

```python
zone_score.plot(x="zone", y="score", kind="bar", figsize=(12,5))
```

2. Growth vs Efficiency Scatter

```python
import seaborn as sns

sns.scatterplot(
    data=zone_score,
    x="efficiency",
    y="growth",
    size="gap",
    hue="zone"
)
```

```python
# Style
sns.set(style="whitegrid")

# Bigger figure
plt.figure(figsize=(12, 8))

# Scatter plot
scatter = sns.scatterplot(
    data=zone_score,
    x="efficiency",
    y="growth",
    size="gap",
    hue="zone",
    palette="tab10",   # better color separation
    sizes=(50, 300),   # control bubble sizes
    alpha=0.8
)

# Move legend outside
plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    borderaxespad=0
)

# Titles and labels
plt.title("Zone Performance: Efficiency vs Growth", fontsize=16, weight='bold')
plt.xlabel("Efficiency", fontsize=12)
plt.ylabel("Growth", fontsize=12)

plt.tight_layout()
plt.show()
```

## Final Story
This project analyzes property tax data from Chennai between 2013 and 2018 to evaluate how effectively the municipal corporation collects revenue across different zones. By comparing the expected tax (demand) with the actual amount collected, a key performance metric—collection efficiency—was derived to measure system effectiveness. The analysis reveals significant variation across zones: some areas consistently achieve near-optimal efficiency, indicating strong compliance and accurate tax estimation, while others exhibit low efficiency, highlighting substantial revenue leakage. Additionally, multiple zones show efficiency values greater than one, suggesting the presence of arrears recovery, penalties, or inconsistencies in demand estimation. Through data cleaning, transformation, exploratory analysis, and feature engineering, the project identifies high-risk zones, uncovers systemic inefficiencies, and provides actionable insights to improve tax collection strategies. Overall, the findings emphasize the need for better demand estimation, targeted enforcement in underperforming areas, and more transparent accounting practices to enhance municipal revenue management.
